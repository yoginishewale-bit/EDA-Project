```python
# Checking information about data.
print(df.info())
def metadata_matrix(df) : 
    return pd.DataFrame({
                'Datatype' : df.dtypes.astype(str), 
                'Non_Null_Count': df.count(axis = 0).astype(int), 
                'Null_Count': df.isnull().sum().astype(int), 
                'Null_Percentage': round(df.isnull().sum()/len(df) * 100 , 2), 
                'Unique_Values_Count': df.nunique().astype(int) 
                 }).sort_values(by='Null_Percentage', ascending=False)

metadata_matrix(df)
```

### Univariate Selection
In this, Statistical tests can be used to select the independent features which have the strongest relationship with the dependent feature. SelectKBest method can be used with a suite of different statistical tests to select a specific number of features.



SelectKBest is used with the chi2 test to select the top 2 features from the iris dataset based on their chi-squared scores

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.datasets import load_iris
```

```python
# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Select the top 2 features using chi-squared test
k_best = SelectKBest(score_func=chi2, k=2)
X_new = k_best.fit_transform(X, y)

# Get the selected features
selected_features = k_best.get_support(indices=True)

print("Selected Features:", selected_features)
```

Univariate Feature Selection:

Use statistical tests to select the top k features based on their individual relationship with the target variable.


```python
from sklearn.feature_selection import SelectKBest, f_classif

# Example: Select top k features using ANOVA F-statistic
selector = SelectKBest(f_classif, k='all')
X_selected = selector.fit_transform(X, y)
```

```python
X_selected
```

```python
from sklearn.ensemble import RandomForestClassifier

# Example: Feature Importance from Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X, y)
feature_importances = rf_model.feature_importances_
```

Recursive Feature Elimination (RFE):

Use RFE to recursively eliminate less important features based on model performance.


```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Example: Recursive Feature Elimination with Random Forest
estimator = RandomForestClassifier()
selector = RFE(estimator, n_features_to_select=5)
X_selected = selector.fit_transform(X, y)
```

Variance Threshold:

Remove features with low variance, as they may not contribute much information.


```python
from sklearn.feature_selection import VarianceThreshold

# Example: Removing low-variance features
selector = VarianceThreshold(threshold=0.1)
X_selected = selector.fit_transform(X)
```

```python
X_selected
```

Data Splitting:

Split your dataset into training and testing sets. This is crucial for assessing the model's performance on unseen data.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

Data Scaling:

Scale numerical features if needed. Many machine learning algorithms perform better when features are on a similar scale.


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Handling Imbalanced Data (if applicable):

If your dataset has imbalanced classes (e.g., significantly more approved loans than denied ones), consider using techniques like oversampling, undersampling, or using algorithms that handle imbalanced data well.


```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)
```

Model Training:

Train your machine learning model using the preprocessed and feature-selected training data.


```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Example: Random Forest Classifier
rf_model = RandomForestClassifier()
param_grid = {'n_estimators': [100, 200, 300], 'max_depth': [10, 20, 30]}
grid_search = GridSearchCV(rf_model, param_grid, cv=5)
grid_search.fit(X_train_scaled, y_train)
```

Model Evaluation:

Evaluate your model's performance on the testing set using appropriate metrics such as accuracy, precision, recall, F1-score, or area under the ROC curve (AUC-ROC).


```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = grid_search.predict(X_test_scaled)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

Binning:

Convert numerical values into bins to reduce the impact of extreme values.


```python
bins = [0, 5000, 10000, 15000, np.inf]
labels = ['0-5000', '5000-10000', '10000-15000', '15000+']
df['CreditScore_bin'] = pd.cut(df['CreditScore'], bins=bins, labels=labels)
```

Robust Scaling:

Use robust scaling that is less sensitive to outliers.


```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df['CreditScore_scaled'] = scaler.fit_transform(df[['CreditScore']])
```
